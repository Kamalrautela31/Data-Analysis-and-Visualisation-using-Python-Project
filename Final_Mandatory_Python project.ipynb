# 6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.

```python
def calculate_median(data):
    """
    Calculate the median value of a dataset.

    Parameters:
        data (list or numpy array): The dataset for which the median needs to be calculated.

    Returns:
        float: The median value of the dataset.
    """

    # Sort the data in ascending order
    sorted_data = sorted(data)

    # Get the number of data points
    n = len(sorted_data)

    # Calculate the index of the median value
    if n % 2 == 0:
        # If the number of data points is even, take the average of the two middle values
        median_index1 = n // 2 + 1
        median_index2 = n // 2
        median = (sorted_data[median_index1] + sorted_data[median_index2]) / 2
    else:
        # If the number of data points is odd, take the middle value
        median_index = n+1 // 2
        median = sorted_data[median_index]

    return median
```

```python
# let us check the median value of housing_median_age column using the calculate_median() function
```

```python
median_value = calculate_median(df['housing_median_age'])
print("Median:", median_value)
```

### 10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.

```python
def categorize_bedroom_size(num_bedrooms):
    if num_bedrooms <= 10:
        return 'small'
    elif num_bedrooms < 1000:
        return 'medium'
    else:
        return 'large'

# Apply the function to the 'total_bedrooms' column and create the new column 'total_bedroom_size'
df['total_bedroom_size'] = df['total_bedrooms'].apply(categorize_bedroom_size)

# Display the updated DataFrame
print(df)
```
